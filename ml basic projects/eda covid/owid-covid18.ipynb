```python
data_cleaned = data.iloc[:,:-5]
```

```python
data_cleaned.drop('handwashing_facilities',axis = 1)
```

```python
data_cleaned.isnull().sum()
```

```python
data_cleaned = data_cleaned.drop(['female_smokers','male_smokers','handwashing_facilities'],axis = 1)
```

```python
data_cleaned
```

```python
numeric_cols = data_cleaned.select_dtypes(include='number').columns
```

```python
data_cleaned[numeric_cols] = data_cleaned[numeric_cols].fillna(data_cleaned[numeric_cols].mean())
```

```python
data_cleaned
```

```python
data_cleaned = data_cleaned.drop_duplicates()
```

```python
data_cleaned.shape
```

```python
selected_columns = [
    'location', 'date', 'total_cases', 'new_cases', 'total_deaths',
    'new_deaths', 'people_vaccinated', 'people_fully_vaccinated',
    'total_boosters', 'population'
]
df_cleaned = data_cleaned[selected_columns]
```

```python
df_cleaned
```

```python
import seaborn as sns
latest = df_cleaned[df_cleaned['date'] == df_cleaned['date'].max()]
top5 = latest.sort_values(by='total_cases', ascending=False).head(5)
print("Top 5 Countries by Total Cases:")
print(top5[['location', 'total_cases']])

# Plot Total Cases of Top 5 Countries
plt.figure(figsize=(10,5))
sns.barplot(data=top5, x='location', y='total_cases')
plt.title("Top 5 Countries by Total COVID-19 Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line Plot: Cases Over Time for One Country
india = df_cleaned[df_cleaned['location'] == 'India']
plt.figure(figsize=(10,5))
plt.plot(india['date'], india['total_cases'], label='Total Cases')
plt.plot(india['date'], india['total_deaths'], label='Total Deaths')
plt.title("COVID Cases & Deaths in India Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
